# Resultaten


````python
# Hier de data en de analyse
temp_begin = 23.2 # graden celcius
temp_einde = 26.8 # graden celcius
delta_temp = temp_einde - temp_begin
spanning = 15.4 # volt
stroom = 1.59  # ampere
tijd =  300 # seconde
massa = 0.4 #kg 
vermogen = spanning * stroom

c = (vermogen*tijd)/(massa*delta_temp)

print(c) 
````
